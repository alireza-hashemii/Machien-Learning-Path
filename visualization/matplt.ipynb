```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(-4,4, 100)
fx = x ** 2
plt.plot(x, fx)
plt.title("A polynomial function graph")
plt.xlabel("x")
plt.ylabel('f(x)= x ^ 2')
```

```python
plt.subplot(1,2,2)
plt.plot(x, 1/x ** 2)
plt.plot(x, np.sin(x))
```

```python
x_1 = np.linspace(0, 5, 10)
y_1 = x_1 ** 2

fig_1 = plt.figure(figsize=(5, 4), dpi=100)
axes_1 = fig_1.add_axes([0.4, .1, .9, .9])


axes_1.plot(x_1, y_1, label="x/ x^2")
axes_1.plot(y_1, x_1, label='x^2/x')

axes_1.legend(loc=2)
```

```python
# figsize = (width, height)
fig_2 = plt.figure(figsize=(6, 4))
axes_test = fig_2.add_axes([0,0,1,1])
axes_test.plot(x_1, y_1, 
               c='navy',
               lw=2, 
               ls='-.', 
               marker='o', 
               markersize=7, 
               markerfacecolor='y',
               markeredgecolor='y'
        )
axes_test.set_xlim([0, 3])
axes_test.set_ylim([0, 10])

axes_test.grid(True, dashes=(5,2,5,2)) 
axes_test.set_facecolor('#FAEDB7')
```
